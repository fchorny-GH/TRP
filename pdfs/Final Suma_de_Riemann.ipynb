# **La suma de Riemann**
Es el nombre que recibe el cálculo aproximado de una integral definida, mediante una sumatoria discreta con un número de términos finito. Una aplicación común es la aproximación del área de funciones en un gráfico.

Fue el matemático alemán Georg Friedrich Bernhard Riemann (1826-1866) quien ofreció por primera vez una definición rigurosa de la integral de una función en un intervalo dado. La dio a conocer en un artículo publicado en 1854.  

<br>  

Suma-de-Riemann-01.jpg

La suma de Riemann se define sobre una función $y = f(x)$, con $x$ perteneciente al intervalo cerrado [a, b]. Sobre este intervalo se efectúa una partición P de n elementos:

$$P = \lbrace{x_0=a, x_1, x_2, ..., x_n=b \rbrace}$$

Esto significa que el intervalo se divide de la siguiente manera:

$$S(f, n) = \sum_{k = 1}^{n} f(t_k)(x_k - x_{k-1})$$
<br>
Aquí $t_k$ está comprendido entre $x_{k-1}$ y $x_k$: $x_{k-1} \leq t_k \leq x_k$

En la figura 1 se muestra gráficamente la suma de Riemann de la función f en el intervalo $[x_0, x_4]$ sobre una partición de cuatro subintervalos, los rectángulos de color gris.

La suma representa el área total de los rectángulos y el resultado de esta suma se aproxima numéricamente al área bajo la curva $f$, entre las abscisas $x=x_0$ y $x=x_4$.

Desde luego, la aproximación al área bajo la curva mejora muchísimo en la medida que el número n de particiones sea mayor. De esta manera la suma converge al área bajo la curva, cuando el número $n$ de particiones tiende a infinito.

# Fórmulas y propiedades

La suma de Riemann de la función f(x) sobre la partición:

$$P = \lbrace{x_0=a, x_1, x_2, ..., x_n=b \rbrace}$$

Definida sobre el intervalo $[a, b]$, está dada por:

$$S(P, f) = \sum_{k = 1}^{n} f(t_k)(x_k - x_{k-1})$$

Donde $tk$ es un valor en el intervalo $[x_k, x_{k-1}]$. En la suma de Riemann se suelen usar intervalos regulares de ancho $\Delta x = \frac{\left( b - a \right)}{n}$, donde $a$ y $b$ son los valores mínimo y máximo de la abscisa, mientras que $n$ es el número de subdivisiones.

Suma inferior queda definida como:

$$S_{inf}(f, P) = \sum_{k=1}^{n} m_k \cdot (x_k - x_{k-1})$$

Donde en cada subintervalo $[x_{k-1}, x_k]$, se considera el *mínimo valor* que toma la funcion $f(x)$:

$$m_{k} = \min_{x \in \left[x_{k-1}, x_k\right]} f(x)$$

# **Propuesta 1** Suma Inferior:

```python
def sumaInferior(funcion, desde, hasta, rectangulos):
    if rectangulos <= 0:
        raise ValueError("En la primaria no te enseñaron que no podes dividir algo por cero?")

    if hasta <= desde:
        raise ValueError("Estas en tu casa y para ir a la Uni, vas desde la Uni a tu Casa?")

    a = desde
    b = hasta
    n = rectangulos
    delta = (b - a) / n
    sumatoria = 0.0

    for k in range(n):
        izq = a + k * delta
        der = izq + delta

        m_k = min(funcion(izq), funcion(der))

        sumatoria += m_k * delta

    return sumatoria
```

# **Propuesta 1 v3** Suma Inferior:

En lugar de mostrar con un print el resultado final, mostrar cada iteración en una tabla, mediante el uso de Plotly Dash.

En el código original solo nos interesaba replicar GeoGebra, es decir, ingresamos los parámetros a la función y que nos dé la solución para los $n$ rectángulos en un intervalo [a, b].

Ahora nos interesa conocer cada cálculo intermedio al que llegó, no sólo la última iteración, estos valores debemos guardarlos para ser utilizados por esto, debemos modificar un poco lo realizado.

```python
import math

entradas = {
    "sen": "math.sin",
    "sin": "math.sin",
    "cos": "math.cos",
    "tan": "math.tan",
    "e^": "math.exp",
    "ln": "math.log",
    "log": "math.log10",
    "x^": "x**"
}

def sanitizar(sintaxis):
    for clave, valor in entradas.items():
        sintaxis = sintaxis.replace(clave, valor)
    return sintaxis

def sumaInferior(funcion, desde, hasta, rectangulos):
    if rectangulos <= 0:
        raise ValueError("En la primaria no te enseñaron que no podes dividir algo por cero?")

    if hasta <= desde:
        raise ValueError("Estas en tu casa y para ir a la Uni, vas desde la Uni a tu Casa?")

    a = desde
    b = hasta
    n = rectangulos
    delta = (b - a) / n
    sumatoria = 0.0
    data = [] # El almacén. Nueva línea!

    for k in range(n):
        izq = a + k * delta
        der = izq + delta

        # m_k = min(funcion(izq), funcion(der))
        """
        min(a,b) Nos oculta el valor de x ganador asi que toca reevaluar,
        antes no nos importaba este dato, solo saber cual era el valor mínimo al evaluar x por izquierda o por derecha, ahora si nos interesa.
        Es por eso, que debemos desahcer la eficacia de min(valor a, valor b) por un ternario o un bucle if.
        """
        # Nuevas líneas
        f_izq = funcion(izq) # Guardamos el f(x) para x izquierda
        f_der = funcion(der) # Guardamos el f(x) para x derecha
        m_k, x_ganador = (f_izq, izq) if f_izq <= f_der else (f_der, der) # Según se cumpla la coindicion central, se asigna (la altura f_izq o f_der en m_k, x ganador izq o der en x_ganador)

        sumatoria += m_k * delta

        # Nueva línea
        """
        Agrega a la lista de datos (data)
        - Cada iteración k
        - El valor del x que fue mínimo al evaluar f(izq) y f(der)
        - El mínimo valor f(x) guardado en m_k
        - El delta x.
        - El área parcial.
        - La suma inferior parcial en cada iteración.
        """

        data.append({"Iteración": k + 1, "x evaluado": round(x_ganador, 4), "f(x)=(x evaluado)^2": round(m_k, 4), "Delta x": round(delta, 4), "Área parcial=f(x) * Delta x": round(m_k * delta, 6), "Suma inferior hasta el momento = S_n + s_(n-1)": round(sumatoria, 6)}) # Para agregar al almacen

    return sumatoria, data # Al almacen! Se agrega data como lo que retorna nuestra función.
```

# **Objetivo Extra:**

Otro desafío: una función SumaSupSumaInf(función, mínimo, máximo, eps)
que haga esto:
eps es un número chiquito. Por ejemplo eps=0.001.
La función hace lo siguiente: busca un número de rectángulos N, tal que sumaSuperior-sumaInferior, para ese N, sea menor que eps.
Y después imprime, desde k=1 hasta k=N este texto



*   $\text{sumaInferior(k) <Integral < sumaSuperior(k)}$

quedan N líneas de texto, donde las dos columnas se van acercando, hasta que su diferencia es menor que eps.

# **Propuesta 3 v1:** Objetivo Extra

```python
import math

entradas = {
    "sen": "math.sin",
    "sin": "math.sin",
    "cos": "math.cos",
    "tan": "math.tan",
    "e^": "math.exp",
    "ln": "math.log",
    "log": "math.log10",
    "x^": "x**"
}

def sanitizar(sintaxis):
    for clave, valor in entradas.items():
        sintaxis = sintaxis.replace(clave, valor)
    return sintaxis

def sumaInferior(funcion, desde, hasta, rectangulos):
    if rectangulos <= 0:
        raise ValueError("En la primaria no te enseñaron que no podes dividir algo por cero?")

    if hasta <= desde:
        raise ValueError("Estas en tu casa y para ir a la Uni, vas desde la Uni a tu Casa?")

    a = desde
    b = hasta
    n = rectangulos
    delta = (b - a) / n
    sumatoria = 0.0

    for k in range(n):
        izq = a + k * delta
        der = izq + delta

        m_k = min(funcion(izq), funcion(der))

        sumatoria += m_k * delta

    return sumatoria

def sumaSuperior(funcion, desde, hasta, rectangulos):
    if rectangulos <= 0:
        raise ValueError("En la primaria no te enseñaron que no podes dividir algo por cero?")

    if hasta <= desde:
        raise ValueError("Estas en tu casa y para ir a la Uni, vas desde la Uni a tu Casa?")

    a = desde
    b = hasta
    n = rectangulos
    delta = (b - a) / n
    sumatoria = 0.0

    for k in range(n):
        izq = a + k * delta
        der = izq + delta

        M_k = max(funcion(izq), funcion(der))

        sumatoria += M_k * delta

    return sumatoria

def SumaSupSumaInf(funcion, minimo, maximo, eps):
    N = 1 # Rectángulos iniciales
    while True:
        inf = sumaInferior(funcion, minimo, maximo, N)
        sup = sumaSuperior(funcion, minimo, maximo, N)
        integral = (inf + sup) / 2 # Nueva línea
        diferencia = abs(sup - inf)

        print(f"{N:3d}: {inf:.6f} < {integral:.6f} < {sup:.6f}") # Cambiado el texto literal por una variable

        if diferencia < eps:
            print(f"\n→ Con N = {N}, la diferencia {diferencia:.6f} < ε = {eps}")
            break
        N += 1 # Evalua para N y en esta linea le suma 1, para que en la siguiente iteración N=2 y asi sucesivamente hasta romper el bucle
```

**Ejecución:**

```python
def main():
    print("Aproximación de la integral con suma superior e inferior")

    ingreso = input("Ingrese la función f(x): ")
    f = sanitizar(ingreso)
    funcion = lambda x: eval(f, {"x": x, "math": math})

    minimo = float(input("Desde (a): "))
    maximo = float(input("Hasta (b): "))
    eps = float(input("Epsilon (tolerancia, ej: 0.01): "))

    print("\nSuma inferior y superior por número de rectángulos:\n")
    SumaSupSumaInf(funcion, minimo, maximo, eps)

main()
```

```python
import math

entradas = {
    "sen": "math.sin",
    "sin": "math.sin",
    "cos": "math.cos",
    "tan": "math.tan",
    "e^": "math.exp",
    "ln": "math.log",
    "log": "math.log10",
    "x^": "x**"
}

def sanitizar(sintaxis):
    for clave, valor in entradas.items():
        sintaxis = sintaxis.replace(clave, valor)
    return sintaxis

def sumaInferior(funcion, desde, hasta, rectangulos):
    if rectangulos <= 0:
        raise ValueError("En la primaria no te enseñaron que no podes dividir algo por cero?")

    if hasta <= desde:
        raise ValueError("Estas en tu casa y para ir a la Uni, vas desde la Uni a tu Casa?")

    a = desde
    b = hasta
    n = rectangulos
    delta = (b - a) / n
    sumatoria = 0.0

    for k in range(n):
        izq = a + k * delta
        der = izq + delta

        m_k = min(funcion(izq), funcion(der))

        sumatoria += m_k * delta

    return sumatoria

def sumaSuperior(funcion, desde, hasta, rectangulos):
    if rectangulos <= 0:
        raise ValueError("En la primaria no te enseñaron que no podes dividir algo por cero?")

    if hasta <= desde:
        raise ValueError("Estas en tu casa y para ir a la Uni, vas desde la Uni a tu Casa?")

    a = desde
    b = hasta
    n = rectangulos
    delta = (b - a) / n
    sumatoria = 0.0

    for k in range(n):
        izq = a + k * delta
        der = izq + delta

        M_k = max(funcion(izq), funcion(der))

        sumatoria += M_k * delta

    return sumatoria

def SumaSupSumaInf(funcion, minimo, maximo, eps):
    N = 1 # Rectángulos iniciales
    data = [] # El almacén. Nueva línea!
    while True:
        inf = sumaInferior(funcion, minimo, maximo, N)
        sup = sumaSuperior(funcion, minimo, maximo, N)
        integral = (inf + sup) / 2
        diferencia = abs(sup - inf)

        #print(f"{N:3d}: {inf:.6f} < Integral < {sup:.6f}") reemplazada por tabla
        data.append({"Rectángulos":N, "Suma Inferior":inf, "Integral": round(integral, 6), "Suma Superior":sup, "Diferencia":diferencia}) # Nueva línea, duda con la Integral, lo tome literal en el anterior

        if diferencia < eps:
            print(f"\n→ Con N = {N}, la diferencia {diferencia:.6f} < ε = {eps}")
            break
        N += 1 # Evalua para N y en esta linea le suma 1, para que en la siguiente iteración N=2 y asi sucesivamente hasta romper el bucle
    return data # Nueva línea, ahora guardamos datos que queremos en la tabla, asi que la función debe devolverlo
```

**Crear tabla**:

Requiere correr la celda con !pip install

```python
import dash
from dash import html
import dash_bootstrap_components as dbc
import pandas as pd
import math

# Nos saltamos la introducción del usuario, para que no parezca complejo el paso
def f(x):
    return x**2 # Ponemos una función en formato Python

datitos = SumaSupSumaInf(f, 0, 2, 0.1) # Llamamos a la función

datos= pd.DataFrame(datitos) # Data Frame de Pandas Cambia de lista a diccionario con la forma {columna:[filas]} y no requiere declarar columns(palabra reservada)

app = dash.Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP]) # Creación de la Dash app con estilo bootstrap

tabla = dbc.Table.from_dataframe(datos, striped = True, bordered = True, hover = True) # Tabla ahora con columnas y filas alimentadas por la funcion

app.layout = html.Div([tabla]) # La web dash app con su contenido una tabla con columnas y filas.

if __name__ == '__main__':
    app.run(port="8052", debug=0) # Levantar web server para ejecutar la app
```

```python
import dash
from dash import html, dcc
import dash_bootstrap_components as dbc
import plotly.graph_objects as go
import numpy as np
import pandas as pd

def f(x):
    return x**2 # Ponemos una función en formato que Python entienda

datitos = SumaSupSumaInf(f, 0, 2, 0.1)
datos= pd.DataFrame(datitos)

# Usamos las columnas del DataFrame para definir los valores de x e y
val_x = datos["Rectángulos"]
val_y_inf = datos["Suma Inferior"]
val_y_sup = datos["Suma Superior"]
val_y_int = datos["Integral"]

colores = ["blue"] * (len(val_x) - 1) + ["orange"]
trazo = go.Figure() # Creamos un objeto gráfico llamado trazo para guardar los datos y estilos del gráfico
# Inferior                             Opcional: mode="markers" para solo ver puntos
trazo.add_trace(go.Scatter(x=val_x, y=val_y_inf, mode="lines+markers", marker=dict(color=colores, size=8), name="Suma inferior")) # Agregamos los x e y unidos por lineas y le asignamos un nombre
# Integral
trazo.add_trace(go.Scatter(x=val_x, y=val_y_int, mode="lines+markers", marker=dict(color="red", size=8), name="Integral")) # Agregamos los x e y unidos por lineas y le asignamos un nombre
# Superior
trazo.add_trace(go.Scatter(x=val_x, y=val_y_sup, mode="lines+markers", marker=dict(color="green", size=8), name="Suma superior")) # Agregamos los x e y unidos por lineas y le asignamos un nombre

# Maquillaje
trazo.update_layout(
    title="Convergencia de las Sumas de Riemann",
    xaxis_title="Número de rectángulos (N)",
    yaxis_title="Suma",
    legend_title="Gráficas",
    template="plotly_dark" # Con import plotly.io as pio print(pio.template.keys()) en una celda aislada, podremos saber las opciones disponibles para template
)

app = dash.Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP]) # Creación de la Dash app con estilo bootstrap

grafico = dcc.Graph(id = "grafico_plantilla", figure=trazo) # grafico pasando figura

app.layout = html.Div([grafico]) # La web dash app con su contenido un grafico

if __name__ == "__main__":
    app.run(port="8053", debug=False) # Levantar web server para ejecutar la app
```
